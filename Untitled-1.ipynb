Теория: 

.env — это способ хранить конфигурацию отдельно от кода.

Что вообще входит в конфигурацию?
Конфигурация — это всё, что может отличаться между окружениями, но не относится к логике программы.
Что обычно входит в конфигурацию API-проекта:
	1.	BASE_URL
	2.	TOKEN
	3.	API_VERSION
	4.	TIMEOUT
	5.	RETRY_COUNT
	6.	Логи (вкл/выкл debug)
	7.	Пути к файлам
	8.	Флаг окружения (prod/stage/dev)

Prod:
BASE_URL=https://api.pokemonbattle.ru
API_VERSION=v2
TIMEOUT=5

BASE_URL=https://stage-api.pokemonbattle.ru
API_VERSION=v2
TIMEOUT=10





```python
from api.client import ApiClient
from config.settings import BASE_URL, TOKEN

client = ApiClient(BASE_URL, TOKEN)

print("Object state:", client.__dict__)

response = client.get("/me")

print("URL sent:", response.request.url)
print("Headers sent:", response.request.headers)
print("Status:", response.status_code)
print("Body:", response.text)
```
